```python
from pyspark.sql import SparkSession
from pyspark.sql import Row
from pyspark.sql.types import StructType, StructField, StringType, IntegerType

# Create SparkSession
spark = SparkSession.builder.appName("JoinsExample").getOrCreate()

# Define schema for DataFrame 1
schema1 = StructType([
    StructField("id", IntegerType(), True),
    StructField("name", StringType(), True)
])

# Define schema for DataFrame 2
schema2 = StructType([
    StructField("id", IntegerType(), True),
    StructField("age", IntegerType(), True)
])

# Create a list of Rows for DataFrame 1
row_list1 = [
    Row(id=1, name="James"),
    Row(id=2, name="Ann"),
    Row(id=3, name="John")
]

# Create a list of Rows for DataFrame 2
row_list2 = [
    Row(id=1, age=23),
    Row(id=2, age=40),
    Row(id=4, age=30)
]

# Create DataFrames
df1 = spark.createDataFrame(row_list1, schema1)
df2 = spark.createDataFrame(row_list2, schema2)

# # Show DataFrames
# df1.show()
# df2.show()

# Perform join with additional condition
result = df1.join(df2, (df1.id == df2.id) & (df2.age > 25))

# Show the result
result.show()

# Perform Inner Join
inner_join_df = df1.join(df2, df1.id == df2.id, "inner")
print("Inner Join")
inner_join_df.show()

# Perform Left Join
left_join_df = df1.join(df2, df1.id == df2.id, "left")
print("Left Join")
left_join_df.show(truncate=False)

# Perform Right Join
right_join_df = df1.join(df2, df1.id == df2.id, "right")
print("Right Join")
right_join_df.show(truncate=False)

# Perform Full Outer Join
full_outer_join_df = df1.join(df2, df1.id == df2.id, "outer")
print("Full Outer Join")
full_outer_join_df.show(truncate=False)
```
