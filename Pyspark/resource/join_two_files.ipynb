```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import lit
from pyspark.sql.types import StructType, StructField, StringType, IntegerType

# Initialize SparkSession
spark = SparkSession.builder.master("local").appName('Modes of DataFrameReader').getOrCreate()

# Define schema for File 1
schema1 = StructType([
    StructField("Name", StringType(), True),
    StructField("Age", IntegerType(), True)
])

# Define schema for File 2
schema2 = StructType([
    StructField("Name", StringType(), True),
    StructField("Age", IntegerType(), True),
    StructField("Gender", StringType(), True)
])

# Create dataset for File 1
data1 = [
    ("Azarudeen, Shahul", 25),
    ("Michel, Clarke", 26),
    ("Virat, Kohli", 28),
    ("Andrew, Simond", 37)
]

# Create dataset for File 2
data2 = [
    ("Rabindra, Tagore", 32, "Male"),
    ("Madona, Laure", 59, "Female"),
    ("Flintoff, David", 12, "Male"),
    ("Ammie, James", 20, "Female")
]

# Create DataFrames from datasets
df1 = spark.createDataFrame(data1, schema1)
df2 = spark.createDataFrame(data2, schema2)

# Add missing column to df1
# df1 = df1.withColumn("Gender", lit(None).cast(StringType()))
# If allow missing columns is set to True, the missing columns will be added to the DataFrame with null values.

# Union the DataFrames
df_combined = df1.unionByName(df2, allowMissingColumns=True)


# Show the combined DataFrame
df_combined.show()
```
