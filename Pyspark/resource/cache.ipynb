```python
from pyspark.sql import SparkSession
from pyspark.sql import Row
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, FloatType
from pyspark import StorageLevel

# Create SparkSession
spark = SparkSession.builder.appName("CachingExample").getOrCreate()

# Define schema
schema = StructType([
    StructField("name", StringType(), True),
    StructField("age", IntegerType(), True),
    StructField("price", FloatType(), True),
    StructField("rooms", IntegerType(), True)
])

# Create a list of Rows
row_list = [
    Row(name="James", age=23, price=300000.0, rooms=3),
    Row(name="Ann", age=40, price=400000.0, rooms=2),
    Row(name="John", age=45, price=500000.0, rooms=4),
    Row(name="Mike", age=30, price=350000.0, rooms=3)
]

# Create DataFrame
df = spark.createDataFrame(row_list, schema)
df.show()

# Cache DataFrame in MEMORY_ONLY
df.cache()
df.show()

# Cache DataFrame in MEMORY_AND_DISK
df.persist(StorageLevel.MEMORY_AND_DISK)
df.show()

# Cache DataFrame in DISK_ONLY
df.persist(StorageLevel.DISK_ONLY)
df.show()

# Unpersist DataFrame
df.unpersist()
df.show()

# Print schema to verify the changes
df.printSchema()
```
